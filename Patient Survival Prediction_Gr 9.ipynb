```python
# H0: Gender and hospital death are independent
# H1: Gender and hospital death are dependent

table = pd.crosstab(df['gender'], df['hospital_death'])
obs_values = table.values

st.chi2_contingency(obs_values)

# As the pvalue is 0.95 > 0.05 (alpha), we fail to reject H0.
# Therefore, the attributes, gender and hospital death are independent.
# Both males and females are at an equal risk of surviving or not surviving any
# particular medical condition.
```

```python
# H0: mean_pre_icu_loss_days(death) <= mean_pre_icu_loss_days(survived)
# H1: mean_pre_icu_loss_days(death) > mean_pre_icu_loss_days(survived)

A = df[df['hospital_death'] == 1]['pre_icu_los_days']
B = df[df['hospital_death'] == 0]['pre_icu_los_days']

stests.ztest(A,B, alternative = 'larger')

# Here, p-value < alpha(0.05), therefore, we reject H0.
# The cases resulting in death are having a higher value of pre_icu_loss_days.
```

```python
# H0: Ethnicity and hospital death are independent
# H1: Ethnicity and hospital death are dependent

table = pd.crosstab(df['ethnicity'], df['hospital_death'])
obs_values = table.values
st.chi2_contingency(obs_values)

# Here, pvalue = 0.0017 < alpha (0.05), we reject H0.
# Therefore, ethnicity and hospital death are dependent.
```

```python
# H0: ICU Admit Source and hospital death are independent
# H1: ICU Admit Source and hospital death are dependent

table = pd.crosstab(df['icu_admit_source'], df['hospital_death'])
obs_values = table.values
st.chi2_contingency(obs_values)

# Here, pvalue < alpha (0.05), we reject H0.
# Therefore, ICU Admit Source and hospital death are dependent.
```

```python
# H0: ICU type and hospital death are independent
# H1: ICU type and hospital death are dependent

table = pd.crosstab(df['icu_type'], df['hospital_death'])
obs_values = table.values
st.chi2_contingency(obs_values)

# Here, pvalue < alpha (0.05), we reject H0.
# Therefore, ICU type and hospital death are dependent.
```

```python
# Capping - Outlier treatment
```

```python
for col in con_col:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3-q1
    ll = q1-(1.5*iqr)
    ul = q3+(1.5*iqr)
    for ind in df[col].index:
        if df.loc[ind,col]>ul:
            df.loc[ind,col]=ul
        elif df.loc[ind,col]<ll:
            df.loc[ind,col]=ll
        else:
            pass
```

```python
y_test_prob = logreg.predict_proba(X_test)
w_fn = 2 # (more importance to FN)
w_fp = 1
cost_func_lst =[]
for th in np.arange(0.1,1,0.1):
    y_test_pred = [0 if pval<th else 1 for pval in y_test_prob[:,1]]
    con_mat = confusion_matrix(y_test,y_test_pred)
    fp = con_mat[0,1]
    fn = con_mat[1,0]
    cost_func = (w_fn*fn)+(w_fp*fp)
    cost_func_lst.append(cost_func)
    
cf_df = pd.DataFrame({'Threshold': np.arange(0.1,1,0.1),'Cost_Func':cost_func_lst})
cf_df[cf_df['Cost_Func']==min(cf_df['Cost_Func'])]
```
